```python
words = []
flag = 0
with open('elonmusk.txt', 'r') as file:
    for line in file:
        line = line.strip() #remove any '\n' or extra blank spaces

        #Remove empty string from list
        if line:
            words.extend(line.split())

        string = str(line)
        string.rstrip()
        string.replace("\'\n\'", ' ')
        string.replace("'\r'", ' ')   
        newWords = string.split()
        words = words + newWords

print('Reference text has',len(words),'words')
print('Reference text: \n', words)
```

```python
chain = {}      #set up a chain to store
n_words = len(words)    #figure out how many words there are
for i, key in enumerate(words):   #for a key in a dictionary
    if n_words > (i + 1):         #if this isn't the last word
        word = words[i + 1]       #get the next word
        if key not in chain:      #if the word isn't in the chain
            chain[key] = [word]   #make an entry and store the next word
        else:
            chain[key].append(word) #append the next word to the entry for that word
print('There are',len(chain),'distinct words')
```

```python
for i in range(10):
    w1 = rand.choice(words)
    # w1 = "Tesla"    #pick a word
    tweet = w1        #start the tweet with that word

    while len(tweet) < 140:    #while you are within the limit
        w2 = rand.choice(chain[w1])  #make a random choice of potential next words from all of the possible next words
        tweet += ' ' + w2   #add that to the tweet (with a space)
        w1 = w2             #make this into the full tweet

    #capitalize first letter of sentence
    tweet = tweet.replace(tweet[0], tweet[0].upper(), 1)
    print(f'Tweet {i}:', tweet, '\n')
```

```python
#now do exactly the same thing, but for the next TWO words

chain = {}
n_words = len(words)
for i, key1 in enumerate(words):
    if n_words > i + 2:      #make sure this isn't the second last word
        key2 = words[i + 1]  #new key
        word = words[i + 2]  #and again
        if (key1, key2) not in chain:  #if the first two words aren't in the chain
            chain[(key1, key2)] = [word]  #make an entry with this word
        else:
            chain[(key1, key2)].append(word)  #add it to the entry
```

```python
for i in range(10):
    r = rand.randint(0, len(words) - 1)
    key = (words[r], words[r + 1])
    tweet = key[0] + ' ' + key[1]

    while len(tweet) < 140:
        w = rand.choice(chain[key])
        tweet += ' ' + w
        key = (key[1], w)

    tweet = tweet.replace(tweet[0], tweet[0].upper(), 1) + '.'
    print(f'Tweet #{i}:', tweet, '\n')
```
